```python
#only want USA races ,50k or 50mi, 2020
```

```python
df[df['Event distance/length']== '50mi' ]
```

```python
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
#combine all the filter together
```

```python
df2 = df[df['Event distance/length'].isin(['50km','50mi']) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#remove (USA) from event name
```

```python
df2 ['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#clean up athlete age category
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h in athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
#drop some columns: Athlete club , athlete country , Athlete year of birth, Athlete age catogery 
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country','Athlete year of birth','Athlete age category' ], axis = 1)
```

```python
df2.head()
```

```python
#clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()== 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#check dupe
```

```python
df2[df2.duplicated() == True]
```

```python
#reset index
```

```python
df2.reset_index(drop = True)
```

```python
#fix types
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#rename columns 
```

```python
df2 = df2.rename(columns = {'Year of event': 'year', 
                            'Event dates': 'race-day', 'Event name' : 'race-name',
                            'Event distance/length' :'race_length' , 
                            'Event number of finishers' :'race_number_of_finishers' ,
                            'Athlete performance':'athlete_performance' ,
                            'Athlete gender ': 'athlete_gender',
                            'Athlete average speed':'athlete_avg_speed' ,
                            'Athlete ID ': 'athlete_id' ,
                            
                            
                           })
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3 = df2[['race-day', 'race-name', 'race_length', 'race_number_of_finishers', 'Athlete ID' ,'Athlete gender', 'athlete_age','athlete_performance', 'athlete_avg_speed']]
```
